```python
# install library pandas and numpy
from google_play_scraper import app

import pandas as pd

import numpy as np
```

```python
# Scraping 2,000 most recent user reviews from the IKD application, ensuring the data is up to date.
from google_play_scraper import Sort, reviews

result, continuation_token = reviews(
    'gov.dukcapil.mobile_id', # App URL
    lang='id',                # Languange = id
    country='id',             # Country = Indonesia
    sort=Sort.NEWEST,         # Newest Data
    count=2000,               # 2000 Data
    filter_score_with=None    # score with 1-5
)
```

```python
# Converting the scraped review data into a structured DataFrame by extracting each attribute (e.g., username, rating, review content) into separate columns for further analysis.
df_ikd = pd.DataFrame(np.array(result),columns=['reviews'])

df_ikd = df_ikd.join(pd.DataFrame(df_ikd.pop('reviews').tolist()))

df_ikd.head()
```
